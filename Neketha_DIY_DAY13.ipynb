```python
import pandas as pd
df1=pd.DataFrame({'name':['Shubham','Adin','Raghul'],
                  'age':['23','28','30']})
display(df1)
df2=pd.DataFrame({'name':['Amrit','Shubham','Ali'],
                  'age':['20','23','28']})
display(df2)
```

```python
df3=pd.concat([df1,df2]).drop_duplicates()
df3=df3.reset_index(drop=True)
display(df3)
```

#Combining dataframes

```python
import pandas as pd
df1=pd.DataFrame({'employee':['Bob','Jake','Lisa','Sue','John','Daisy'],
                  'department':['Accounting','Engineering','Engineering','HR','Engineering','Engineering']})
df2=pd.DataFrame({'employee':['Lisa','Bob','Jake','Sue','Daisy','john'],
                  'joinning_date':[2004,2008,2012,2022,2021,2020]})
display(df1,df2)
df3=pd.merge(df1,df2)
```

```python
df3=pd.merge(df1,df2)
display(df3)
```
