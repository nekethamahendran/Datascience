# Data Grouping-I

```python
import pandas as pd
import numpy as np
df=pd.read_csv("/content/big_bang_theory_episodes.csv").fillna(np.nan)
df
```

```python
df.groupby(by='directed_by').count()
```

# Printing the first group from grouped DataFrame:

```python
result=df.groupby('directed_by')
display(result.first())
```

# Displaying the group related to director-Mark Cendrowski:

```python
result.get_group('Mark Cendrowski')
```

# Data Grouping-II

```python
import pandas as pd
import numpy as np
df=pd.read_csv("/content/big_bang_theory_episodes.csv")
display(df)
```

# Printing the average viewers in the United States and the  director and the writer's details

```python
df.groupby(by=["directed_by",'written_by']).count()['episode_num_in_season']
```

```python
result=df.groupby(["directed_by",'written_by']) ['us_viewers'].mean()
result
```
